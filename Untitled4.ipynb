# **RBCD approach**

In this project, we are going to work on the RBBD method, which is introduced in the paper "Random block coordinate descent methods for computing optimal transport and convergence analysis". The general optimal transport in a discrete form canbe written as follows:

$\int_{\pi \in \Pi(\mu,\nu)}c(x,y)d\pi(x,y)$

where $c(x,y)$ is the cost function for transporting mass from the point $x$ in the source domain $X$ to the point $y$ in the target domain $Y$. $\pi$ is called the transportation plan from point $x\in X$ to point $y \in Y$. 


```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf
import random
from numpy.random import default_rng
import pandas as pd
from scipy.optimize import linprog
from numpy import linalg as LA
```

```python
x=np.random.uniform(low=-1, high=1.0, size=101)
y=np.random.uniform(low=-1, high=1.0, size=101)

r1=(1/1001)*np.ones(101)
r2=(1/1001)*np.ones(101)
input_shape=np.shape(x)[0]
output_shape=np.shape(y)[0]
N=len(x)
plt.plot(x, y)
plt.ylabel('distribution 1 (1D)')
plt.xlabel('distribution 2 (1D)')

plt.show()
plt.plot(r1, r2)
```

```python
def my_cost_function(x, y):
    """
    calculating the cost 
    """
    my_cost=[]
    Aup=np.eye(N)
    Adown=np.block([[np.ones((1,N)),np.zeros((1,N**2-N))]])


    for i in range(N-1):

        Aup=np.block([[Aup,np.eye(N)]])
    Aup=np.reshape(Aup,(N,N**2))
    
    for i in range(len(x)):
      for j in range(len(y)):
          my_cost.append((1/2)*(x[i]-y[j])**2)
    new_cost=np.reshape(my_cost,(len(x),len(y)))
    b=np.append(r1,r2)
    b.reshape(len(x)+len(y),1)
    for i in range(1,N):
      row=np.block([[np.zeros((1,i)),np.ones((1,N)),np.zeros((1,N**2-N-i))]])
      Adown=np.block([[Adown],[row]])
    A=np.block([[Aup],[Adown]])
    return my_cost,b, A
```

```python
xO=[1,0,0,0,1,0,0,0,1]
[c,b,A]= my_cost_function(x, y)
res = linprog(c, A_ub=None, b_ub=None,  A_eq=A, b_eq=b, method ='revised simplex',callback = None, 
                       options = None, x0 = None)
my_solution=res.x
print(res.message)

print(my_solution.reshape(N,N))
#bounds=(0, None))

  
```
